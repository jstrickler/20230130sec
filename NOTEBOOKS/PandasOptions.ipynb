```python
x = pd.describe_option(_print_desc=False)
print(x[:500])
```

```python
import re
rx_settings = re.compile(r"[a-z]+\..*\s:")
```

```python
data = x.splitlines()
data[:60]
settings = []
current_setting = []
for line in data:
    if rx_settings.match(line):
        if current_setting:
            current_setting[0:1] = current_setting[0].split(' : ')
            settings.append(tuple(current_setting))
            current_setting.clear()
    current_setting.append(line.strip())
        
settings[:10]
len(settings)
```
