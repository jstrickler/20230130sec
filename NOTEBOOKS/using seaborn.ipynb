# Using the seaborn library

Note: examples from https://seaborn.pydata.org/introduction.html#:~:text=Seaborn%20is%20a%20library%20for,explore%20and%20understand%20your%20data with some updates and additions.

First, import necessary modules

```python
import pandas as pd
import seaborn as sns
```

Read some data

```python
tips = sns.load_dataset("tips")
tips['percent'] = (tips['tip'] / tips['total_bill']) * 100
tips
```

```python
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    style="sex", size="percent", hue='day', size_norm=(0,30),
)
```
